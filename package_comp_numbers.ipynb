# 42/44

```python
RALL_bayexpress = pd.read_csv('DGE_results/RALL_bayexpress.csv').iloc[:,1:].rename(columns={'genes': "locus_name"})

# we don't loose any genes here for nans
# RALL_bayexpress = RALL_bayexpress.dropna(subset=['BF_21', 'FC'])

# criteria 1 for significance: (BF_21 > 1) & (abs(FC) > 1)
RALL_bayexpress['FC1_bayexpress'] = (RALL_bayexpress.BF_21 > 1) & (abs(RALL_bayexpress.FC) > 1)

RALL_bayexpress
```

```python
RALL_edgeR = pd.read_csv('DGE_results/RALL_edgeR.csv').rename(columns={'genes': "locus_name"})

# we don't loose any genes here for nans
# RALL_edgeR = RALL_edgeR.dropna(subset=['logFC', 'PValue'])

# however, we do have already lost some during their analysis


# criteria 1 for significance: (pvalue < 0.05) & (abs(log2FoldChange) > 1)
RALL_edgeR['FC1_edgeR'] = (RALL_edgeR.PValue < 0.05) & (abs(RALL_edgeR.logFC) > 1)

RALL_edgeR
```

```python
RALL_DESeq2 = pd.read_csv('DGE_results/RALL_DESeq2.csv').rename(columns={'Unnamed: 0': "locus_name"})

# this is a bit crazy but we loose a lot of genes for nans here
RALL_DESeq2 = RALL_DESeq2.dropna(subset=['log2FoldChange', 'pvalue'])

# criteria 1 for significance: (pvalue < 0.05) & (abs(log2FoldChange) > 1)
RALL_DESeq2['FC1_DESeq2'] = (RALL_DESeq2.pvalue < 0.05) & (abs(RALL_DESeq2.log2FoldChange) > 1)

RALL_DESeq2
```

```python
RALL = pd.DataFrame({'locus_name': RALL_bayexpress.locus_name})

# RALL = pd.concat([RALL, RALL_bayexpress.iloc[:,-2:], RALL_edgeR.iloc[:,-2:], RALL_DESeq2.iloc[:,-2:]], axis=1)

RALL = RALL.merge(RALL_bayexpress[['locus_name', 'FC1_bayexpress']], on='locus_name', how='left')

# display(RALL)

RALL = RALL.merge(RALL_edgeR[['locus_name', 'FC1_edgeR']], on='locus_name', how='left')

RALL = RALL.merge(RALL_DESeq2[['locus_name', 'FC1_DESeq2']], on='locus_name', how='left')

# RALL.to_csv('results_RALL.csv')



RALL
```

```python
# counts (watch the NaNs!)

display(RALL.FC1_bayexpress.value_counts(dropna=False))

display(RALL.FC1_edgeR.value_counts(dropna=False))

display(RALL.FC1_DESeq2.value_counts(dropna=False))
```

# 20/20

```python
# all bayexpress results
# 20 replicates

R20_bayexpress = pd.read_csv(f'DGE_results/R20_1_bayexpress.csv').iloc[:,1:2]

R20_bayexpress_FC1 = pd.read_csv(f'DGE_results/R20_1_bayexpress.csv').iloc[:,1:2]
R20_bayexpress_BF_21 = pd.read_csv(f'DGE_results/R20_1_bayexpress.csv').iloc[:,1:2]


for i in range(1,101):

    data = pd.read_csv(f'DGE_results/R20_{i}_bayexpress.csv').iloc[:,1:]

    # criteria 1 for significance: (BF_21 > 1) & (abs(FC) > 1)
    R20_bayexpress_FC1[f'{i}'] = (data.BF_21 > 1) & (abs(data.FC) > 1)

    R20_bayexpress_BF_21[f'{i}'] = (data.BF_21 > 1)

    R20_bayexpress[f'BF_21{i}'] = data.BF_21
    R20_bayexpress[f'FC{i}'] = data.FC

display(R20_bayexpress_FC1)
display(R20_bayexpress_BF_21)
display(R20_bayexpress)

# R20_bayexpress_FC1.set_index('locus_name').transpose().describe().loc['freq']

display(R20_bayexpress_FC1.set_index('locus_name').transpose().describe())

R20_bayexpress_FC1_True = [(R20_bayexpress_FC1.iloc[i].value_counts().get(True, 0)) for i in range(len(R20_bayexpress_FC1))]
R20_bayexpress_BF_21_True = [(R20_bayexpress_BF_21.iloc[i].value_counts().get(True, 0)) for i in range(len(R20_bayexpress_BF_21))]
```

```python
# all edgeR results
# 20 replicates


R20_edgeR_FC1 = pd.read_csv(f'DGE_results/R20_1_edgeR.csv').iloc[:,:1]

for i in range(1,101):

    data = pd.read_csv(f'DGE_results/R20_{i}_edgeR.csv')

    # criteria 1 for significance: (pvalue < 0.05) & (abs(log2FoldChange) > 1)
    R20_edgeR_FC1[f'{i}'] = (data.PValue < 0.05) & (abs(data.logFC) > 1)


R20_edgeR_FC1['#R20'] = [(R20_edgeR_FC1.iloc[i].value_counts().get(True, 0)) for i in range(len(R20_edgeR_FC1))]

display(R20_edgeR_FC1)

display(R20_edgeR_FC1.set_index('genes').transpose().describe())
```

```python
# 20 replicates
# all DESeq2 results

R20_DESeq2_FC1 = pd.read_csv(f'DGE_results/R20_1_DESeq2.csv').iloc[:,0:1].rename(columns={'Unnamed: 0': "genes"})


for i in range(1,101):

    data = pd.read_csv(f'DGE_results/R20_{i}_DESeq2.csv')

    # criteria 1 for significance: (pvalue < 0.05) & (abs(log2FoldChange) > 1)
    R20_DESeq2_FC1[f'{i}'] = (data.pvalue < 0.05) & (abs(data.log2FoldChange) > 1)


R20_DESeq2_FC1['#R20'] = [(R20_DESeq2_FC1.iloc[i].value_counts().get(True, 0)) for i in range(len(R20_DESeq2_FC1))]

display(R20_DESeq2_FC1)

display(R20_DESeq2_FC1.set_index('genes').transpose().describe())
```
